# Low Pass Filters: 1st Order, Passive
- Series LR
- Series RC


```python
# Series LR LP Filter
#-- Given R = 1kOhm, F_c = 1kHz, Plot H(jw)

# Vin o--[L]--+--o Vo
#             |
#            [R]
#             |
#            GND

R = 1000
F_c = 1000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#L = R/w_c
L = R / (2*np.pi*F_c)

R_L = R/L

# From circuit solution: 
# Mag = R/L / (jw + R/L)
# Find phase using angle command
H_s = R_L / (R_L + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series LR LP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

```python
# Series RC LP Filter
#-- Given R = 10kOhm, F_c = 8kHz, Plot H(jw)

# Vin o--[R]--+--o Vo
#             |
#            [C]
#             |
#            GND

R = 10000
F_c = 8000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#C = 1/(w_c*R)
C = 1 / (R*2*np.pi*F_c)

RC = R*C

# From circuit solution: 
# Mag = R/L / (jw + R/L)
# Find phase using angle command
H_s = (1/RC) / ((1/RC) + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series RC LP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

# High Pass Filters: 1st Order, Passive
- Series CR
- Series RL


```python
# Series CR HP Filter
#-- Given R = 10kOhm, F_c = 8kHz, Plot H(jw)

# Vin o--[C]--+--o Vo
#             |
#            [R]
#             |
#            GND

R = 10000
F_c = 8000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#C = 1/(w_c*R)
C = 1 / (R*2*np.pi*F_c)

RC = R*C

# From circuit solution: 
# Mag = jw / (jw + 1/RC)
# Find phase using angle command
H_s = (i*w) / ((1/RC) + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series CR HP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

```python
# Series RL HP Filter
#-- Given R = 10kOhm, F_c = 10kHz, Plot H(jw)

# Vin o--[R]--+--o Vo
#             |
#            [L]
#             |
#            GND

R = 10000
F_c = 10000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#L = R/(w_c)
L = R / (2*np.pi*F_c)

R_L = R/L

# From circuit solution: 
# Mag = jw / (jw + 1/RC)
# Find phase using angle command
H_s = (i*w) / (R/L + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series LR HP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

```python
# Series RL HP Filter with parallel load
#-- Given R = 10kOhm, F_c = 10kHz, Plot H(jw)

# Vin o--[R]--+----+---o Vo
#             |    |
#            [L]  [R]
#             |    |
#            GND   GND

R = 10000
F_c = 10000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#L = R/(w_c)
L = R / (2*np.pi*F_c)

R_L = R/L

K = R/(R+R) # for refrence if R1 =/= R2

# From circuit solution: 
# Mag = jw / (jw + 1/RC)
# Find phase using angle command
H_s = (i*w)*K / (K*R/L + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series LR HP Filter with Parallel Load Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

# Band-Pass Filters: 

Center Frequency: $\omega_0 = \sqrt{\omega_\text{c1} \omega_\text{c2}}$

Bandwidth: $\beta = \omega_\text{c2} - \omega_\text{c1}$

Quality Factor: $Q = \frac{\omega_0}{\beta}$


**Series RLC:** $H(s) = \frac{\frac{R}{L}s}{s^2 + \frac{R}{L}s + \frac{1}{LC}}$ , 
$\omega_0 = \sqrt{\frac{1}{LC}}$ , $\omega_\text{c1,c2} = \pm \frac{R}{2L} + \sqrt{(\frac{R}{2L})^2 + (\frac{1}{LC})}$ , $\beta = \frac{R}{L}$ , $Q = \sqrt{\frac{L}{CR^2}} = \frac{f_0}{f_2 - f_1}$

```python
# Series RLC BP Filter
#-- Select 1kHz to 10kHz:
f_1 = 1e3
f_2 = 10e3
#-- thus center frequency f_0 = \sqrt{f_1 f_2} = 3162 Hz.
f_0 = math.sqrt(f_1 * f_2) 
#-- Choose C = 1 microF
C = 1e-6 

#-- Thus we find the Inductor value:
L = 1/(C * 2*np.pi*f_0 * 2*np.pi*f_0)       #~= 2.5 mH

#-- And the Qulaity Factor:
Q = f_0 / (f_2 - f_1)                     # ~= 0.35

#-- And finally find the Resistor value:
R = math.sqrt(L / (C*Q*Q))                #~= 143 Ohms

f = np.arange(0, 1000000, 50)             # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

i_w = i*w
R_L = R/L

# From circuit solution: 
# Find phase using angle command
H_s = (i_w)*(R_L) / (- w*w + R_L*i_w + 1/(L*C))

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    markers_on = [int(f_1 / 50), int(f_2 / 50)]
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series RLC BP Filter Magnitude/Phase vs Frequency')

    ax1.semilogx(f,H_s_mag, '-ob', markevery=markers_on)
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')
    
    # Annotate passband:
    ax1.text(f_1 * 0.8, 0.7, "f_1", ha='right')
    ax1.text(f_2 * 0.8, 0.7, "f_2", ha='right')
    ax1.text(f_0, 0.1, "PASSBAND", ha='center')
    ax1.axvline(x=f_1, color='b', linestyle='--')
    ax1.axvline(x=f_2, color='b', linestyle='--')
    
    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

# Band-Reject Filters
Similar to Band-Pass, but with magnitude = 0 at w_0 (rather than magnitude = 1). 

**Series RLC:** $H(s) = \frac{ s^2 + \frac{1}{LC}}{s^2 + \frac{R}{L}s + \frac{1}{LC}}$ , 
$\omega_0 = \sqrt{\frac{1}{LC}}$ , $\omega_\text{c1,c2} = \pm \frac{R}{2L} + \sqrt{(\frac{R}{2L})^2 + (\frac{1}{LC})}$ ,
$\beta = \frac{R}{L}$ , $Q = \sqrt{\frac{L}{CR^2}} = \frac{f_0}{f_2 - f_1}$

```python
# Series RLC BS Filter
#-- Reject with center frquency of 750Hz, and bandwidth of 250Hz. Use a Capacitor with 100 nF value.
f_0 = 750
beta = 250
f_1 = f_0 - beta
f_2 = f_0 + beta
w_0 = 2*np.pi*f_0

C = 1e-9

#-- Thus we find the Inductor value:
L = 1/(C * w_0*w_0)                       # = 450 mH

#-- And the Qulaity Factor:
Q = w_0 / (beta)                          # = 3

#-- And finally find the Resistor value:
R = beta * L                              #~= 707 Ohms

f = np.arange(0, 1000000, 50)             # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

i_w = i*w
R_L = R/L
_LC = 1/(L*C)

# From circuit solution: 
# Find phase using angle command
H_s = ((- w*w) + _LC) / (- w*w + R_L*i_w + _LC) 

# Alternative method (more useful for more complicated Transfer Functions):
#-- Use scipy.signal.freqs, build filter using iirfilter or firfilter functions

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    markers_on = [int(f_1 / 50), int(f_2 / 50)]
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series RLC BS Filter Magnitude/Phase vs Frequency')

    ax1.semilogx(f,H_s_mag, '-ob', markevery=markers_on)
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')
    
    # Annotate passband:
    ax1.text(f_1*0.9, 0.95, "f_1", ha='right')
    ax1.text(f_2*1.5, 0.95, "f_2", ha='right')
    ax1.text(f_0, 0, "STOPBAND", ha='center')
    ax1.axvline(x=f_1, color='b', linestyle='--')
    ax1.axvline(x=f_2, color='b', linestyle='--')
    
    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

# Parallel (RLC) Resonance Ciruit
**Parallel RLC:** $H(j\omega) = \frac{1}{1 + j(\omega R C - \frac{R}{\omega L})}$ , 
$\omega_0 = \sqrt{\frac{1}{LC}}$ , $\omega_\text{c1,c2} = \pm \frac{R}{2L} + \sqrt{(\frac{R}{2L})^2 + (\frac{1}{LC})}$ ,
$\beta = \frac{R}{L}$ , $Q = \sqrt{\frac{L}{CR^2}} = \frac{f_0}{f_2 - f_1} = \frac{\omega_0}{\beta}$

```python
# Parallel RLC as Band-Pass Filter:
#-- Center frequency 1kHz, Bandwidth 750Hz, using inductor value of 10 mH.
f_0 = 1e3
beta = 750
f_1 = f_0 - beta
f_2 = f_0 + beta
w_0 = 2*np.pi*f_0

L = 10e-3

#-- Thus we find the Capacitor value:
C = 1/(L * w_0*w_0)                       # = 2.53 microF

#-- And the Qulaity Factor:
Q = w_0 / (beta)                          # = 1.33

#-- And finally find the Resistor value:
R = 1/ (beta * C)                         #~= 84 Ohms

f = np.arange(0, 1000000, 50)             # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

i_w = i*w

# From circuit solution: 
# Find phase using angle command
H_s = 1 / ( 1 + i_w*C*R - i*(R/(w*L)))    # w in denominator will cause div by 0 

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    markers_on = [int(f_1 / 50), int(f_2 / 50)]
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Parallel RLC BP Filter Magnitude/Phase vs Frequency')

    ax1.semilogx(f,H_s_mag, '-ob', markevery=markers_on)
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')
    
    # Annotate passband:
    ax1.text(f_1 * 0.8, 0.7, "f_1", ha='right')
    ax1.text(f_2 * 1.5, 0.7, "f_2", ha='right')
    ax1.text(f_0, 0.1, "PASSBAND", ha='center')
    ax1.axvline(x=f_1, color='b', linestyle='--')
    ax1.axvline(x=f_2, color='b', linestyle='--')
    
    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

# Contour Networks: 
### From: https://www.diyaudioandvideo.com/Calculator/ContourNetwork/Help/
Used to correct rising frequency response. Goes between input and filter.

$f : \text{frequency to start correcting response.}$
$f_m : \text{frequency at which response peaks.}$
$A_t : \text{attenuation from f to f_m.}$
$R_d : \text{total driver impedance.}$

- **RL**: used when response rises when frequency increases.

$$L = \frac{0.15916}{f}$$

$$X = 2 \pi f_m L$$

- **RC**: used when response rises when frequency decreases.

$$C = \frac{0.15916}{f}$$

$$X = \frac{1}{2 \pi f_m L}$$

Then the total impedance is found:

$$Z = \frac{R X}{\sqrt{R^2 + X^2}}  , R = \sqrt{\frac{X^2}{\frac{X^2}{R^2} - 1}}$$

And attenuation of the contour network:

$$A_t = 20*log_{10} \frac{R_d + Z}{R_d} ,  Z = R_d * 10^{\frac{A_t}{20}} - R_d$$
