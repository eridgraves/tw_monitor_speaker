```python
# Modeling Filter Frequency Responses
import numpy as np
from matplotlib import pyplot as plt
```

# Low Pass Filters:
- Series LR
- Series RC


```python
# Series LR LP Filter
#-- Given R = 1kOhm, F_c = 1kHz, Plot H(jw)

R = 1000
F_c = 1000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#L = R/w_c
L = R / (2*np.pi*F_c)

R_L = R/L

# From circuit solution: 
# Mag = R/L / (jw + R/L)
# Find phase using angle command
H_s = R_L / (R_L + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series LR LP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```

```python
# Series RC LP Filter
#-- Given R = 10kOhm, F_c = 8kHz, Plot H(jw)

R = 10000
F_c = 8000

f = np.arange(0, 1000000, 25)               # Set up frequency space
w = 2*np.pi*f

i = np.array([1j], dtype=complex)         # The imaginary number

#C = 1/(w_c*R)
C = 1 / (R*2*np.pi*F_c)

RC = R*C

# From circuit solution: 
# Mag = R/L / (jw + R/L)
# Find phase using angle command
H_s = (1/RC) / ((1/RC) + i*w)

H_s_mag = abs(H_s)
H_s_phase = np.angle(H_s, deg=True)

# Plot Magnitude and Phase responses
with plt.style.context('ggplot'): # Set style to be like R
    
    fig, ax1 = plt.subplots(figsize=(10,5)) 
    
    plt.title('Series RC LP Filter Magnitude vs Frequency')

    ax1.semilogx(f,H_s_mag, 'b')
    ax1.set_ylabel('Filter Magnitude (dB)',color='blue') 
    ax1.set_xlabel('Frequency (Hz)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.semilogx(f,H_s_phase)
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Filter Phase Angle (Degrees)',color='r');
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
```
